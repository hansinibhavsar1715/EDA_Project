## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Cleaning Reviews
df['Reviews'] = df['Reviews'].astype(str)
df['Reviews'] = df['Reviews'].replace('[M]', '000000', regex=True)
df['Reviews'] = df['Reviews'].replace('[K]', '000', regex=True)
df['Reviews'] = df['Reviews'].replace('[.]', '', regex=True)
df['Reviews'] = df['Reviews'].replace('', '0')
df['Reviews'] = df['Reviews'].astype(int)

# Cleaning Installs
df['Installs'] = df['Installs'].astype(str)
df['Installs'] = df['Installs'].str.replace('+','')
df['Installs'] = df['Installs'].str.replace(',','')
df['Installs'] = df['Installs'].astype(int)

# Cleaning Price (FINAL FIX)
df['Price'] = df['Price'].astype(str)
df['Price'] = df['Price'].replace('Free', '0')
df['Price'] = df['Price'].str.replace('$','')
df['Price'] = df['Price'].astype(float)

# Handling Missing Values
df['Rating'].fillna(df['Rating'].mean(), inplace=True)

df.head()


````

### What all manipulations have you done and insights you found?

Data cleaning and preprocessing were performed by correcting data types, removing special characters, and handling missing values. Columns like Reviews, Installs, and Price were converted into numerical formats. Missing values in Rating were replaced with the mean. These steps made the dataset ready for analysis and visualization.
